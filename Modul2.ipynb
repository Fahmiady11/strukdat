Dengan menggunakan **modul stack** yang sudah ada, buatlah fungsi-fungsi tambahan untuk mengoperasikan ekspressi matematika postfix, dengan syarat input adalah ekspressi matematika infix (termasuk fungsi untuk pengecekan kurung).  Oleh karena itu buatlah tiga buah fungsi berikut :
1. isValidate : yaitu untuk pengecekan kurung dari ekspresi matematika infix.  Return value adalah boolean (True jika tidak ada kurung yang salah, dan False jika sebaliknya)
2. in2Post : yaitu konversi ekspressi matematika infix menjadi ekspressi matematika postfix (ingat, operand dapat berupa angka dengan satu digit, dua digit, dan seterusnya, dapat juga berupa spasi)
3. evaluatePost : yaitu operasi matematika dengan inputan berupa ekspressi postfix

Main code untuk mengeksekusi fungsi-fungsi yang sudah dibuat adalah :
```
strInfix=input('Masukkan Ekspressi Matematika - Infix =')
if isValidate(strInfix):
    strPostFix=in2Post(strInfix)
    print('Ekspressi Matematika PostFix = ',strPostFix)
    print('Hasil Evaluasi = ',evaluatePost(strPostFix))
else:
    print('Tidak dapat dioperasikan')            
    
```
Berikut adalah ekspressi matematika infix yang harus ditest kedalam fungsi-fungsi yang sudah dibuat :
1. [(5+6)*(20-14)
2. [(5+6)*(20-14)]
3. {[23+56]*[12-2]:[10+20]]
4. {[23+56]*[12-2]:[10+20]}
5. {[23+56] *[12- 2]:[10 +20]}
6. {[23+56]*[12-2]:10
7. {[23+56]*[12-2]}:10
8. 12 : { 90-5}*3
9. 12:{90-5}*3
