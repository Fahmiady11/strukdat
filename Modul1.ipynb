Buatlah suatu fungsi yang dapat mengoperasikan fungsi matematika berikut ini :
\begin{eqnarray}
y&=&\frac{\sum_{x=1}^{n}(3x+4)}{5n}
\end{eqnarray}


Misalkan :
$n=1$, 
\begin{eqnarray}
y&=&\frac{\sum_{x=1}^{1}(3x+4)}{5\times 1}\\
&=&\frac{3+4}{5}
\end{eqnarray}

$n=3$, 
\begin{eqnarray}
y&=&\frac{\sum_{x=1}^{3}(3x+4)}{5\times 3}\\
&=&\frac{7+10+13}{15}\\
&=&\frac{30}{15}\\
&=&2
\end{eqnarray}
